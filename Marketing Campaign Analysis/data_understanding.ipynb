DATA IMPORT:

```python
# To import the dataset

import pandas as pd
import numpy as np

df = pd.read_csv("../Datasets/marketing_campaign_data.csv")
df.head()
```

```python
df.info()
```

DATA CLEANING:

```python
# To convert Dt_customer from object to DateTime format

df["Dt_Customer"] = pd.to_datetime(df["Dt_Customer"])
df[["Dt_Customer"]].head()
df["Dt_Customer"].dtype
```

```python
# To find the age of customer

current_year = 2026
df['Age'] = current_year - df["Year_Birth"]
df.head()
```

```python
# To find the children count per family

df['Children_count'] = df['Kidhome'] + df['Teenhome']
df.head()
```

```python
# To find the total spends per family/customer

df["Total_Spend"] = df["MntWines"] + df["MntFruits"] + df["MntMeatProducts"] + df["MntFishProducts"] + df["MntSweetProducts"] + df["MntGoldProds"]
df.head()
```

```python
# To find the total purchases

df["Total_Purchases"] = df["NumWebPurchases"] + df["NumCatalogPurchases"] + df["NumStorePurchases"] + df["NumDealsPurchases"]
df.head()
```

```python
reference_date = df['Dt_Customer'].max()
reference_date
```

```python
# To get Customer tenure

df["Customer_Tenure_Days"] = (reference_date - df["Dt_Customer"]).dt.days
# the above is to get customer tenure in days
df["Customer_Tenure_Months"] = (df["Customer_Tenure_Days"] / 30).round(1)
# the above is to get customer tenure in months

df.head()
```

```python
df.describe()
```

```python
df.info()
```

```python
df.shape
```

```python
# To check age outliers

df[df["Age"] < 18].shape
df[df["Age"] > 90].shape
```

```python
# To check income outliers - there could be no income less than 0

df[df["Income"] <= 0].shape
```

```python
# To check total spend outliers - there could be no spend in negative value

df[df["Total_Spend"] < 0].shape
```

```python
# to remove outliers

df = df[(df["Age"] >= 18) & (df["Age"] <= 90)]

df = df[df["Income"] > 0]

df.shape

# Not usig IQR technique here as IQR would remove top contributors (top 5 - 10% customers)
# Not suitable for Business data, as IQR is more statistical.
```

EXPLORATORY DATA ANALYSIS: BIVARIATE/MULTIVARIATE ANALYSIS:

```python
# To perform bivariate analysis for Response vs Income

import matplotlib.pyplot as plt

response_income = df.groupby("Response")["Income"].mean()

plt.figure()
response_income.plot(kind="bar")
plt.title("Income by Response")
plt.xlabel("Response (0 = No, 1 = Yes)")
plt.ylabel("Income")
plt.show()
```

```python
# To perform bivariate analysis for Response vs Age

response_age = df.groupby("Response")["Age"].mean()

plt.figure()
response_age.plot(kind="bar")
plt.title("Age by Response")
plt.xlabel("Response (0 = No, 1 = Yes)")
plt.ylabel("Age")
plt.xticks(rotation=0)
plt.show()
```

```python
# To perform bivariate analysis for Response vs Total Spend

response_spend = df.groupby("Response")["Total_Spend"].mean()

plt.figure()
response_spend.plot(kind="bar")
plt.title("Total Spend by Response")
plt.xlabel("Response (0 = No, 1 = Yes)")
plt.ylabel("Total Spend")
plt.xticks(rotation=0)
plt.show()
```

```python
# To perform bivariate analysis for Response vs Product Spend

product_cols = [
    "MntWines",
    "MntFruits",
    "MntMeatProducts",
    "MntFishProducts",
    "MntSweetProducts",
    "MntGoldProds"
]

product_response = df.groupby("Response")[product_cols].mean()

product_response.T.plot(kind="bar")
plt.title("Product Spend by Response")
plt.xlabel("Product Category")
plt.ylabel("Product Spend")
plt.xticks(rotation=45)
plt.show()
```

```python
# To perform bivariate analysis for Response vs Channel Usage

channel_cols = [
    "NumWebPurchases",
    "NumCatalogPurchases",
    "NumStorePurchases",
    "NumDealsPurchases"
]

channel_response = df.groupby("Response")[channel_cols].mean()

channel_response.T.plot(kind="bar")
plt.title("Channel Usage by Response")
plt.xlabel("Channel")
plt.ylabel("Total Purchases")
plt.xticks(rotation=45)
plt.show()
```

```python
# To perform bivariate analysis for Campaign Acceptance vs Customer Count

plt.bar(campaign_acceptance.index, campaign_acceptance.values)
plt.title('Campaign Acceptance Count')
plt.xlabel('Campaigns')
plt.ylabel('Customer Count')
plt.xticks(rotation=45)
plt.show()
```

Key Insights and Patterns:

3. Spending on premium product categories such as wines, meat products, and gold products is noticeably higher among campaign responders.

RULE BASED SEGMENTATION:


```python
# Segmentation for High Income

df["High_Income"] = (df["Income"] > 75000).astype(int)

df.groupby("High_Income")["Response"].mean()
```

```python
# Segmentation for Young Customers

df["Young_Customer"] = (df["Age"] < 30).astype(int)

df.groupby("Young_Customer")["Response"].mean()
```

```python
# Segmentation for Campaign Responders

df["Campaign_Responder"] = df["Response"]

df.groupby("Campaign_Responder")["Response"].mean()
```

```python
# Segmentation for High Web Engagement

df["High_Web_Engagement"] = (df["NumWebVisitsMonth"] > 5).astype(int)

df.groupby("High_Web_Engagement")["Response"].mean()
```

```python
# Segmentation for Family Customer

df["Family_Customer"] = (df["Children_count"] > 0).astype(int)

df.groupby("Family_Customer")["Response"].mean()
```

```python
# Segmentation for High Spender

high_spend_threshold = df["Total_Spend"].quantile(0.90)
high_spend_threshold
```

```python
df["High_Spender"] = (df["Total_Spend"] > high_spend_threshold).astype(int)

df.groupby("High_Spender")["Response"].mean()
```

```python
segmented_data = ['High_Income', 'Young_Customer', 'Campaign_Responder', 'High_Web_Engagement', 'Family_Customer', 'High_Spender']
df[segmented_data].head()
```

```python
# Customers in each segment

segment_cols = [
    "High_Income",
    "Young_Customer",
    "Campaign_Responder",
    "High_Web_Engagement",
    "Family_Customer",
    "High_Spender"
]

segment_size = df[segment_cols].mean() * 100

segment_size.plot(kind="bar")
plt.title("Customer Distribution by Segment (%)")
plt.ylabel("Percentage of Customers")
plt.xticks(rotation=45)
plt.show()
```
