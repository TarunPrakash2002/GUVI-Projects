```python
# this cell is to extract aircraft data according to flights data
# in this cell we specifically extract the aircraft info from flights file

import requests
import json
import os

aircrafts = set()

airports = [
    "MAA", "BLR", "DEL", "DXB",
    "SYD", "FRA", "JFK", "LHR",
    "MEL", "HKG", "SIN", "DOH"
]

for code in airports:

  flights_path = f"/content/drive/MyDrive/project/data/raw/flights/{code}_combined.json"
  with open(flights_path, "r") as f:
    data = json.load(f)

  # Arrival aircrafts
  for flight in data.get("arrivals", []):
    reg = flight.get("aircraft", {}).get("reg")
    if reg:
      aircrafts.add(reg)

  # Departure aircrafts
  for flight in data.get("departures", []):
    reg = flight.get("aircraft", {}).get("reg")
    if reg:
      aircrafts.add(reg)

print('No. of Unique regs:', len(aircrafts))
print(aircrafts)
```
